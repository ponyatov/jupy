**Jupyter Notebook** – невероятно мощный инструмент для разработки методов, выполнения вычислений, и представления результатов в области интерактивной обработки данных. [В этой статье](/user/ponyatov/notebooks/Welcome.ipynb) вы узнаете, как использовать сервер Jupyter в вашей работе.

**Блокнот** (*notebook*) объединяет документацию, код программ (*скрипты*), и результаты их работы в единый документ, который совмещает в себе визуализацию, текст описания, математические уравнения, графики, элементы графического ввода, и т.д. Использование блолкнота -- интуитивно понятный рабочий процесс, который обеспечивает быструю итеративную разработку программ обработки данных, и готовый отчет об их работе, что делает ноутбуки очень популярным выбором для специалистов по анализу данных, и инженеров различных специализаций.

Для создания **выполняемых ячеек** блокнота может использоваться несколько языков программирования, но самым популярным для работы с данными стал [язык Python](https://pythonworld.ru/samouchitel-python). Это произошло благодаря тому, что Питон (или Пайтон) легко осваивается любым техническим специалистом, не обязательно имеющим какой-то опыт в программировании. Его *синтаксис* (внешний вид программ, их построение) очень легко читается, а огромная библиотека готовых компонентов позволяет делать множество крутых вещей, не слишком сильно закапываясь в технические детали.

## зачем использовать Jupyter

Задача Jupyter-сервера -- служить промежуточным звеном между постановкой задачи, и реализацией законченного специализированного сервиса. В некоторых случаях пользователям можно предоставить возможность самостоятельно работать с блокнотами, менять код, изменять параметры.
* черновой вариант частей сервиса, для обкатки решений и вариантов реализации
* задачи, требующие решения в ручном пошаговом режиме, с изменением параметров
* одноразовые операции (преобразование данных между форматами, загрузки/выгрузки в БД)
* администрирование через API интерфейсы
* быстрое "грязное" решение задачи, предоставляемое пользователям на период разработки и отладки полнофукционального сервиса

## Управление

Так как Jupyter разрабатывался под активный ввод в основном скриптов и команд на Python, в нем реализована вполне удобная система управления с клавиутуры, почти не требующая использования мыши.

### основные сочетания клавиш

|||
-|-
клавиатурные стрелки|перемещение по блокноту, выбор текущей ячейки
`[Ctrl]+[Enter]` | **вычисление текущей ячейки**
`[Shift]+[Enter]`| вычисление ячейки **с автоматическим переходом к следующей**
`[Ctrl]+[Shift]+[-]`|разделение ячейки на две по текущей строке курсора в редакторе
`[Ctrl]+[/]`|(раз)коммектирование выделенного Python-кода в редакторе
`[Ctrl]+[L]`|переформатирование кода по стандарту (`autopep8`)
`[M]`|сменить тип ячейки на **M**arkdown (язык разметки текста)
`[Y]`|сменить тип ячейки на P**y**thon
`[A]`|вставить новую ячейку *выше* (**a**bove)
`[B]`|вставить новую ячейку *ниже* (**b**elow)
`[Enter]`|вход в редактор текущей ячейки
`[Esc]`| выход из редактора (без пересчета ячейки)

Полный список клавиатурных комбинаций можно посмотреть в `меню/Help/Keyboard Shortcuts`
